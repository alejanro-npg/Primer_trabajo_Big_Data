```python
import pandas as pd
import os
from IPython.display import display
```

- We'll define the paths for the files selected for the analysis.



```python
csv_path = 'result_retrieve_left-and-right_x_50_2016_modified.csv'
parquet_path = 'result_retrieve_left-and-right_x_50_2016_modified.parquet'
```

- Now, we'll load the two files.

```python
# --- Load the modified files ---
print("Loading modified files into new DataFrames...")
df_csv = pd.read_csv(csv_path)
df_parquet = pd.read_parquet(parquet_path)

print("Modified files loaded successfully. Here is a preview of the data:")
display(df_csv.head())
```

- Let's compare the disk size and the shape (rows, columns) of the two file formats. You'll notice that Parquet is significantly more efficient for storage.

```python
# --- Compare file size and DataFrame shape ---

# Get file sizes
csv_size_bytes = os.path.getsize(csv_path)
parquet_size_bytes = os.path.getsize(parquet_path)

# Get DataFrame shapes
csv_rows, csv_cols = df_csv.shape
parquet_rows, parquet_cols = df_parquet.shape

# Print comparison
print("--- File and DataFrame Comparison ---")
print("\nCSV File:")
print(f"  - File Path: {csv_path}")
print(f"  - Size on disk: {csv_size_bytes / 1024:.2f} KB")
print(f"  - Shape: {csv_rows} rows, {csv_cols} columns")

print("\nParquet File:")
print(f"  - File Path: {parquet_path}")
print(f"  - Size on disk: {parquet_size_bytes / 1024:.2f} KB")
print(f"  - Shape: {parquet_rows} rows, {parquet_cols} columns")

# Highlight the size difference
size_difference = (csv_size_bytes - parquet_size_bytes) / csv_size_bytes * 100
print(f"\nNote: The Parquet file is {size_difference:.2f}% smaller than the CSV file.")
```

This is the core analysis step. We group the data by the specified categories and calculate the average value_x, value_y, and value_z for each group.

```python
# --- Create average values for x, y, and z columns ---

# Define the columns to group by and the columns to aggregate
grouping_cols = ['fact_id', 'side', 'joint', 'variable']
value_cols = ['value_x', 'value_y', 'value_z']

print(f"Grouping by {grouping_cols} and calculating the mean of {value_cols}...")

# Perform the groupby and aggregation.
# .reset_index() converts the grouped columns back into regular columns.
df_agg = df_parquet.groupby(grouping_cols)[value_cols].mean().reset_index()

# Rename columns for clarity in the database
df_agg.rename(columns={
    'value_x': 'avg_x',
    'value_y': 'avg_y',
    'value_z': 'avg_z'
}, inplace=True)

print("\nPreview of the final data to be loaded:")
display(df_agg.head())
```

### 1. Handling Missing Data

```python
# --- 1. Handling Missing Data ---
print("--- Missing Values Count ---")
missing_counts = df_parquet[['value_x', 'value_y', 'value_z']].isna().sum()
print(missing_counts)
```

```python
# --- Create a cleaned DataFrame without missing values in value_x, value_y, or value_z ---
df_clean = df_parquet.dropna(subset=['value_x', 'value_y', 'value_z'])

print("\n--- Cleaned DataFrame Info ---")
print(f"Original shape: {df_parquet.shape}")
print(f"Cleaned shape: {df_clean.shape}")
```

```python
# --- Verify missing values in the cleaned DataFrame ---
print("\n--- Missing Values in Cleaned DataFrame ---")
print(df_clean[['value_x', 'value_y', 'value_z']].isna().sum())

# --- Remove unnecessary columns ---
cols_to_drop = ['sd_x', 'sd_y', 'sd_z', 'md_x', 'md_y', 'md_z']
df_clean = df_clean.drop(columns=cols_to_drop, errors='ignore')

print("\n--- Cleaned DataFrame after dropping unnecessary columns ---")
print(f"Shape after drop: {df_clean.shape}")
display(df_clean.head())
```

```python
# --- Filter for only 'Hip' joint ---
df_hip = df_clean[df_clean['joint'] == 'Hip']

# Show the number of rows
hip_rows = len(df_hip)

print(f"\nNumber of rows in the 'Hip' DataFrame: {hip_rows}")
display(df_hip.head())
```

### 2. File Format Comparison

```python
# --- 2. File Format Comparison ---

# Guardar en CSV y Parquet
df_hip.to_csv("hip_data.csv", index=False)
df_hip.to_parquet("hip_data.parquet", index=False, engine="pyarrow")
```

```python
# Obtener tamaños de archivo
csv_size = os.path.getsize("hip_data.csv")
parquet_size = os.path.getsize("hip_data.parquet")
```

```python
# Calcular reducción porcentual
percentage_diff = ((csv_size - parquet_size) / csv_size) * 100

# Mostrar resultados
print("--- File Size Comparison ---")
print(f"Tamaño CSV: {csv_size} bytes")
print(f"Tamaño Parquet: {parquet_size} bytes")
print(f"Diferencia: {csv_size - parquet_size} bytes")
print(f"El archivo Parquet es {percentage_diff:.2f}% más pequeño que el CSV")
```

### 3. Advanced Panda Aggregation

```python
# Group by 'side' and 'variable', then calculate std for the three value columns

std_df = (
    df_clean
    .groupby(['side', 'variable'])[['value_x', 'value_y', 'value_z']]
    .std()
    .reset_index()
)

# Display the aggregated dataframe
print("\n--- Aggregated Standard Deviations ---")
print(std_df)
```

```python

# Filter for 'L' (Left) side and find the variable with highest std in value_x

highest_std_var = (
    std_df[std_df['side'] == 'L']
    .loc[:, ['variable', 'value_x']]
    .sort_values(by='value_x', ascending=False)
    .iloc[0]
)
print(f"Variable: {highest_std_var['variable']}, Std(value_x): {highest_std_var['value_x']}")

```

```python
# Índice del valor máximo en value_y
idx_max = df_clean['value_y'].idxmax()

# fact_id correspondiente
fact_id_max = df_clean.loc[idx_max, 'fact_id']

print(f"fact_id: {fact_id_max}, value_y: {idx_max}")
```
